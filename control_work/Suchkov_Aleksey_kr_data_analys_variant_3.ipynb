**Вопрос №1**

Что такое NumPy и зачем он используется в анализе данных?

NumPy это библиотека написанная на языке Pyton для расчётов. Используется для того чтобы удобнее считать однотипные массивы- матрицы. NumPy спроектирован так, что выполняет операции с матрицами не через массивы, а через побитовые операции.
В анализе данных используется для быстрого подсчёта матриц с последующим анализом результатов для выявления каких то выводов.  


**Вопрос №2**

Каковы характеристики и основные типы данных основного объекта массива в NumPy?

Характеристики массивов в NumPy: их элементы должны быть одного типа данных. Если есть вложенные массивы, они должны быть одного размера(формы\shape)
Типы данных этих массивов могут быть int16\32\64\128, float16\32\64\128

**Вопрос №3**

Что такое Matplotlib.PyPlot и для чего он используется при визуализации данных?

Это модуль библиотеки Matplotlib который отвечает за построение графиков. Используется для того чтобы было нагляднее видеть данные и анализировать их.

**Вопрос №4**

Опишите различные типы графиков, которые можно создать с помощью Matplotlib.PyPlot, и объясните, для чего они нужны.

Графики: Plot/линейный, строится для визуализации данных в виде линии. Перепад температуры, изменение плотности населения и т д.

Scatter/точечный, строится для визуализации данных и поиска зависимостей. Рисуется две оси, и множество точек, по этим точкам определяется зависимость одной оси к другой, или не определяется зависимость, потому что рисунок точек сплошное месиво.

Pie/круговая диаграмма- строится для определения частей\долей во всех данных. То есть какая характеристика занимает сколько места из всех характеристик. Представляет собой пиццу\пирог из кусков.

Bar/гистограмма, строится для визуализации показателей. Даёт наглядно понять, какой столбец больше и на сколько. Представляет собой башенки.

**Вопрос №5**

Что такое описательная статистика и какие статистические характеристики она включает?

Описательная статистика это статистика которая описывает данные засчёт показателей которые были получены.
Включает в себя моду; среднее значение; медиану
среднее- среднее арифмитическое всех показаний. Сумма всех элементов поделённая на количество.
мода- самое наиболее часто встречаемое показание
медиана - элемент который посередине в упорядоченных показаниях, если элементов чётное, то медиана это сумма двух средних элементов. не путать с средним арифмитическим. Просто элемент по середине упорядоченного списка.

Создайте линейный график, на котором будет отображено изменение уровней PM2.5 и PM10 по месяцам. Используйте разные цвета для каждого типа частиц.

```python
import pandas as pd
df = pd.read_csv('out.csv')
df
```

```python
import matplotlib.pyplot as plt
```

1. Создайте линейный график, на котором будет отображено изменение уровней PM2.5 и PM10 по месяцам. Используйте разные цвета для каждого типа частиц.

```python
months = df['Месяц']
pm25 = df['Уровень PM2.5']
pm10 = df['Уровень PM10']
plt.plot(months, pm25, color = "purple")
plt.plot(months, pm10, color = "green")
plt.xlabel('Месяц')
plt.ylabel('Уровни загрязнения')
plt.title('Изменение уровней PM2.5 и PM10 по месяцам')
plt.xticks(rotation=45)
plt.legend(labels=['PM2.5', 'PM10'])
plt.show()
```

2. Постройте гистограмму, чтобы показать распределение уровней PM10 по месяцам. Это поможет определить, в каких месяцах чаще всего фиксируются высокие уровни загрязнения.

```python
months = df['Месяц']
pm10 = df['Уровень PM10']
avg_pm10 = pm10.mean()
colors = ["pink" if i <= avg_pm10 else "red" for i in pm10]
plt.bar(months, pm10, color=colors)
plt.xlabel('Месяц')
plt.ylabel('Уровни загрязнения')
plt.title('Изменение уровней PM10 по месяцам')
plt.xticks(rotation=45)
plt.legend(labels=['PM10'])
plt.show()
```

3. Для выбранного города создайте круговую диаграмму, показывающую долю месяцев с уровнем PM2.5 выше среднего по сравнению с месяцами с уровнем PM2.5 ниже среднего.

```python
pm25 = df['Уровень PM2.5']
avg_pm25 = pm25.mean()
more_avg = (pm25 > avg_pm25).sum()
less_avg = (pm25 < avg_pm25).sum()
plt.pie([more_avg, less_avg], labels=['Больше','Меньше'])


plt.title('Доля месяцев меньше и больше среднего значения загрязнения PM2.5')
plt.legend(['Больше','Меньше'])

plt.show()
```
